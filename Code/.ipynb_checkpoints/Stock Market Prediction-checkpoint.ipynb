```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, LSTM, Dropout
```

```python
# Load the training data from CSV file
data = pd.read_csv('Google_train_data.csv')

# Display the first few rows of the dataset
data.head()

# Display information about the dataset (data types, missing values, etc.)
data.info()

# Convert 'Close' column to numeric, forcing any errors to be NaN
data["close"] = pd.to_numeric(data.Close, errors='coerce')

# Drop rows with NaN values to ensure data integrity
data = data.dropna()

# Select only the 'close' price column for training
trainData = data.iloc[:, 4:5].values

# Scale the data to the range [0, 1] for better performance in neural networks
sc = MinMaxScaler(feature_range=(0, 1))
trainData = sc.fit_transform(trainData)
```

```python
# Prepare training data
X_train = []
Y_train = []

# Create sequences of 60 time steps to predict the next time step
for i in range(60, len(trainData)):
    X_train.append(trainData[i-60:i, 0])
    Y_train.append(trainData[i, 0])

# Convert lists to numpy arrays
X_train, Y_train = np.array(X_train), np.array(Y_train)

# Reshape X_train to be 3D (samples, time steps, features)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
# Build the LSTM model
model = Sequential()

# Add an Input layer specifying the input shape
model.add(Input(shape=(X_train.shape[1], 1)))

# Add LSTM layers with dropout for regularization
model.add(LSTM(units=100, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=100, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=100, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=100, return_sequences=False))
model.add(Dropout(0.2))

# Add a Dense layer with a single neuron (output layer)
model.add(Dense(units=1))

# Compile the model using Adam optimizer and mean squared error loss function
model.compile(optimizer='adam', loss='mean_squared_error')

# Print model summary to verify the structure
model.summary()
```

```python
# Train the model on the training data
hist = model.fit(X_train, Y_train, epochs=20, batch_size=32, verbose=2)

# Plot the training loss over epochs
plt.plot(hist.history['loss'])
plt.title('Training model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
# Load the test data from CSV file
testData = pd.read_csv('Google_test_data.csv')

# Convert 'Close' column to numeric, forcing any errors to be NaN
testData["Close"] = pd.to_numeric(testData.Close, errors='coerce')

# Drop rows with NaN values to ensure data integrity
testData = testData.dropna()

# Select only the 'close' price column for testing
testData = testData.iloc[:, 4:5]

# Actual values for testing
Y_test = testData.iloc[60:, 0:].values

# Prepare input data for the model
inputClosing = testData.iloc[:, 0:].values

# Scale the input data using the same scaler as for training data
inputClosing_scaled = sc.transform(inputClosing)

# Prepare X_test
X_test = []
length = len(testData)
timestamp = 60

# Create sequences of 60 time steps for the test data
for i in range(timestamp, length):
    X_test.append(inputClosing_scaled[i-timestamp:i, 0])

# Convert list to numpy array
X_test = np.array(X_test)

# Reshape X_test to be 3D (samples, time steps, features)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Predict stock prices using the trained model
Y_pred = model.predict(X_test)

# Inverse transform to get actual stock prices from the predicted values
predicted_price = sc.inverse_transform(Y_pred)
```

```python
# Plot the actual vs. predicted stock prices
plt.plot(Y_test, color='red', label='Actual Stock Price')
plt.plot(predicted_price, color='green', label='Predicted Stock Price')
plt.title('Google stock price prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
